```python
plt.figure(figsize=(17, 5))
fig, ax = plt.subplots()

diag = plt.bar(data['Region'], data[' Estimated Unemployment Rate (%)'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=60, ha='right')
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.title('Estimated Unemployment Rate by Region')
plt.show()
```

```python
fig = px.line(data, x=' Date', y=' Estimated Unemployment Rate (%)', title='Unemployment Rate Over Time', labels={' Estimated Unemployment Rate (%)': 'Unemployment Rate'}, template='plotly_dark')

# Covid-19 Impact Analysis
covid_events = {'start_date': '2020-03-01', 'end_date': '2021-06-30', 'event': 'Peak of Pandemic'}
fig.add_shape(type='rect', x0=pd.to_datetime(covid_events['start_date']),
              x1=pd.to_datetime(covid_events['end_date']), y0=0, y1=1,
              fillcolor='red', opacity=0.3, layer='below', line=dict(color='red', width=2))
fig.update_layout(xaxis_title='Date', yaxis_title='Unemployment Rate', legend_title='Region')
fig.update_xaxes(type='category')
fig.show()
```

```python
fig = px.bar(data, x='Region', y=' Estimated Unemployment Rate (%)', color='Region',
             animation_frame=' Date', title='Unemployment Rate Over Time',
             labels={'Estimated Unemployment Rate (%)': 'Unemployment Rate (%)'},
             template='plotly',
             color_discrete_sequence=px.colors.sequential.Viridis,
             category_orders={'Region': ['Region1', 'Region2', 'Region3']})

fig.update_layout(xaxis={'categoryorder': 'total descending'},
                  xaxis_title='Region', yaxis_title='Unemployment Rate (%)',
                  barmode='stack',
                  showlegend=False)

fig.add_annotation(text='Impact of Unemployment Over Time',
                   xref='paper', yref='paper',
                   x=0.5, y=1.05,
                   showarrow=False,
                   font=dict(size=16, color='black'))

fig.show()
```
